```python
# 필요한 라이브러리 설치 (최초 한 번만 실행)
!pip install imbalanced-learn keras-tuner shap xgboost -q

# 라이브러리 임포트
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, precision_recall_curve
from imblearn.over_sampling import SMOTE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from keras_tuner.tuners import RandomSearch
import shap
import xgboost as xgb
from google.colab import drive

# 구글 드라이브 마운트
drive.mount('/content/drive')

# 데이터 불러오기
file_path = '/content/drive/MyDrive/data/W_F.csv'
try:
    df = pd.read_csv(file_path, encoding='cp949')
except:
    df = pd.read_csv(file_path, encoding='utf-8-sig')

# 전처리
df['폭염여부'] = df['폭염여부(O/X)'].map({'O': 1, 'X': 0})
df['월'] = pd.to_datetime(df['일시'], errors='coerce').dt.month
df = df[df['월'].isin([5, 6, 7, 8, 9])]
df['일교차'] = df['최고기온(°C)'] - df['최저기온(°C)']
df['습도지수'] = df['평균기온(°C)'] * df['평균상대습도(%)'] / 100

features = ['최고체감온도(°C)', '최고기온(°C)', '평균기온(°C)', '최저기온(°C)', '평균상대습도(%)', '일교차', '습도지수']
X = df[features].apply(pd.to_numeric, errors='coerce')
y = df['폭염여부']
X = X.dropna()
y = y.loc[X.index]

# Stratified K-Fold에서 첫 번째 fold만 사용
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
train_idx, test_idx = next(skf.split(X, y))
X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

# 학습셋에 SMOTE 적용
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# 표준화
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_res)
X_test_scaled = scaler.transform(X_test)

# 모델 정의 함수 (Keras Tuner용)
def build_model(hp):
    model = Sequential()
    model.add(Dense(units=hp.Int('units1', 32, 128, step=32), activation='relu', input_shape=(X_train_scaled.shape[1],)))
    model.add(Dense(units=hp.Int('units2', 16, 64, step=16), activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# 하이퍼파라미터 튜닝
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=1,
    directory='tuner_dir',
    project_name='heatwave'
)
tuner.search(X_train_scaled, y_train_res, epochs=30, validation_split=0.2, callbacks=[EarlyStopping(patience=3)], verbose=0)
best_model = tuner.get_best_models(num_models=1)[0]

# 최적 모델 학습
history = best_model.fit(X_train_scaled, y_train_res, validation_split=0.2, epochs=50, batch_size=16,
                         callbacks=[EarlyStopping(patience=5)], verbose=0)

# 정확도 시각화
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.legend()
plt.title('ANN 정확도')
plt.show()

# 평가
y_pred_prob = best_model.predict(X_test_scaled)
y_pred = (y_pred_prob > 0.5).astype(int)
print("ANN Classification Report:\n", classification_report(y_test, y_pred))
print("ANN ROC-AUC:", roc_auc_score(y_test, y_pred_prob))

# PR Curve
precision, recall, _ = precision_recall_curve(y_test, y_pred_prob)
plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("ANN PR Curve")
plt.show()

# SHAP 해석
explainer = shap.Explainer(best_model.predict, X_train_scaled[:100])
shap_values = explainer(X_train_scaled[:100])
shap.summary_plot(shap_values, X_train_scaled[:100], feature_names=X.columns)

# XGBoost 학습 및 평가
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(X_train_scaled, y_train_res)
xgb_pred = xgb_model.predict(X_test_scaled)
xgb_prob = xgb_model.predict_proba(X_test_scaled)[:, 1]
print("XGBoost Classification Report:\n", classification_report(y_test, xgb_pred))
print("XGBoost ROC-AUC:", roc_auc_score(y_test, xgb_prob))

# ROC Curve 비교
fpr_ann, tpr_ann, _ = roc_curve(y_test, y_pred_prob)
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, xgb_prob)
plt.plot(fpr_ann, tpr_ann, label='ANN')
plt.plot(fpr_xgb, tpr_xgb, label='XGBoost')
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Curve 비교")
plt.legend()
plt.show()

# 앙상블 평가
ensemble_prob = (y_pred_prob.flatten() + xgb_prob) / 2
ensemble_pred = (ensemble_prob > 0.5).astype(int)
print("Ensemble Classification Report:\n", classification_report(y_test, ensemble_pred))
print("Ensemble ROC-AUC:", roc_auc_score(y_test, ensemble_prob))

# 앙상블 PR Curve
precision, recall, _ = precision_recall_curve(y_test, ensemble_prob)
plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Ensemble PR Curve")
plt.show()

# 날짜 기반 사용자 입력 예측 함수
def predict_heatwave(date_str, 최고체감온도, 최고기온, 평균기온, 최저기온, 평균상대습도):
    일교차 = 최고기온 - 최저기온
    습도지수 = 평균기온 * 평균상대습도 / 100
    input_df = pd.DataFrame([{
        '최고체감온도(°C)': 최고체감온도,
        '최고기온(°C)': 최고기온,
        '평균기온(°C)': 평균기온,
        '최저기온(°C)': 최저기온,
        '평균상대습도(%)': 평균상대습도,
        '일교차': 일교차,
        '습도지수': 습도지수
    }])
    input_scaled = scaler.transform(input_df)
    prob = best_model.predict(input_scaled)[0][0]
    result = "폭염 발생" if prob > 0.5 else "폭염 아님"
    print(f"날짜: {date_str}")
    print(f"예측 결과: {result} (확률: {prob:.4f})")

# 예측 함수 호출 (예시 입력)
predict_heatwave("2025-07-14", 최고체감온도=37.5, 최고기온=35.1, 평균기온=29.3, 최저기온=24.0, 평균상대습도=67)
```
